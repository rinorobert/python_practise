## Question 2
Write a Python program to compare two three-digit numbers **A** and **B** and check the following condition:

- Whether the **first digit of A** is less than the **last digit of B**

Use string indexing to extract the required digits, convert them to integers, and print **True** or **False** based on the comparison.

Taking A=123 and B=378

```python
A=int(input())
B=int(input())
A=str(A)
B=str(B)
leng_B=len(B)
print(A[0] < B[leng_B-1])
```

## Question 3
Write a Python program to check whether the **last three characters** of two given strings are the same.

- Read two input strings.
- Extract the last three characters from each string using slicing.
- Compare the extracted parts.
- Print **True** if they match, otherwise print **False**.

Taking a = apple and b = pimple

```python
a=input()
b=input()
leng_a=len(a)
leng_b=len(b)
is_same=a[-3:leng_a] == b[-3:leng_b]
print(is_same)
```

## Question 4
Write a Python program to check whether the **area of a rectangle** is less than or equal to its **perimeter**.

You will be given two inputs:
- Length of the rectangle
- Breadth of the rectangle

Calculate:
- Area = length × breadth
- Perimeter = 2 × (length + breadth)

Print **True** if the area is less than or equal to the perimeter, otherwise print **False**.

Taking lh=3 and bh=4


```python
lh=int(input())
bh=int(input())
area=lh*bh
perimeter=2*(lh+bh)
is_true=area <= perimeter
print(is_true)
```

## Question 5
Write a Python program that reads a **string** and a **number N**, and checks whether the **first N characters** of the string are different from the **last N characters** of the same string.

Print **True** if they are different, otherwise print **False**.

Taking a=toronto and N=2

```python
a=input()
N=int(input())
lh=len(a)
is_true=a[:N] != a[-N:lh]
print(is_true)
```
